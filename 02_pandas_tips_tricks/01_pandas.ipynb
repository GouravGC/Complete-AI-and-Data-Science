# Data Cleaning

```python
# Handling Missing Data
import seaborn as sns
df = sns.load_dataset('penguins')
df.head()
```

```python
df.info()
```

```python
# Finding the missing values in data
df.isnull().sum()
```

```python
# Filling missing values
df['sex']= df['sex'].fillna(df['sex'].mode()[0])
```

```python
# dropping missing values
df.dropna(inplace=True)
```

```python
# Data type conversion
df['body_mass_g'] = df['body_mass_g'].astype('int')
```

```python
# Removing duplicates
df = df.drop_duplicates()
```

# Data Manipulation & Analysis 

```python
# grouping and aggregation tools
print(df.groupby('species')['island'].value_counts())
df.groupby('species')['island'].unique()
```

```python
# Crosstable
from IPython.display import display

display(pd.crosstab(df['species'], df['island']))
display(pd.crosstab(df['species'], df['sex']))
```

```python
df.head()
```

```python
df['sex'] = df['sex'].str.replace('Male','men').astype(object)
```

```python
df['sex'].unique()
```

```python
print(pd.pivot_table(
    data=df,
    index='sex',
    columns='species',
    values='body_mass_g',   # any column (even numeric dummy)
    aggfunc='count'
))

pd.pivot_table(
    data=df,
    index='sex',
    columns='species',
    values='body_mass_g',
    aggfunc='mean'
)
```
